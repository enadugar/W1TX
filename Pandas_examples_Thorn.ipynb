# Pandas examples


## Imports

```python
import pandas as pd
import altair as alt
import datetime
```

## Creating DataFrames

```python
# Create a DataFrame named df1 with four columns and initialize data

df1 = pd.DataFrame({
  'col0': [0, 1, 2],  'col1': [3, 4, 5],
  'col2': ['ab', 'cd', 'ef'],
  'col3': [datetime.datetime.now()] * 3})
```

```python
#Print the DataFrame
df1
```

```python
#Use Pandas to read the csv file named in single quotes, then print the DataFrame
df2=pd.read_csv('https://runestone.academy/ns/books/published/DSA301S22/_static/world_countries.csv')
df2
```

## Inspecting DataFrames

```python
# Only the last one below will display
# To display others, use the function display()
# Or copy and paste in another cell
df1.head()     # First 5 rows
df1.tail()     # Last 5 rows
df1.columns    # Columns names
len(df1)       # Number of rows
df1.shape      # Number of rows and cols
df1.describe() # Stats about each column
df1.info()     # Summary info
```

```python
df1.shape
```

## Summarizing columns

```python
df2.columns
```

```python
# Rename a column
df2 = df2.rename(
  columns={'Population': 'Pop'})
df2.Pop.sum()    # Sum
df2.Pop.mean()   # Average
df2.Pop.std()    # Standard deviation
df2.Pop.median() # Median
df2.Pop.min()    # Minimum
df2.Pop.max()    # Maximum
```

```python
df2.Pop
```

## Filtering rows

```python
df2[5:11]      # Select rows 5 through 10
# Rows with Spain in the Country column
df2[df2.Country == "Spain"]
# Removing nulls
df2 = df2[~df2.Pop.isnull()]
# Convert strings to integers
df2.ConSal = df2.Pop.astype('int64')
# Booleans operators are &, | and ~
df2[(df2.Pop > 100) &
    ~(df2.Area.isnull())]
```

## Column manipulations

```python
# Arithmetic operations on columns
df1['col0'] + df1['col1']
# Even if they're strings
df1['col2'] + df1['col2']
# Create new column from the result
df1['col4'] = df1['col0'] / df1['col1']
# String methods and attributes can be
# accessed via .str.
df1['col2'].str.replace('a', 'b')
# And datetime methods and attributes
# via .dt.
df1.col3.dt.date
# Select just some columns from DataFrame
df2[['Country', 'Pop']]
```

## Dealing with missing values

```python
# Drop rows with any missing values
df2.dropna()
# Drop columns with any missing value
df2.dropna(axis=1)
# Fill missing values with 0s
df2.fillna(0)
# Fill missing values with ''
df2.fillna('')
```

## Grouping

```python
# Get the average salary for each country
df2.groupby('Country').agg(
    {'Pop': 'mean'})
# Get the average and minimum salary
df2.groupby('Country').agg(
    {'Pop': ['mean', 'min']})
# Keep grouping column as a column
df2.groupby(
    'Country', as_index=False).agg(
    {'Pop': ['mean', 'min']})
```

## Miscellaneous

```python
# Reorder from top salary to lowest
df2.sort_values('Pop',
                ascending=False)
# Remove a column
df2.drop(columns='Phones')
# Randomly select a sample of 45 rows
df2 = df2.sample(45)
```

## Merging

```python
df3 = pd.DataFrame(
    {'col5': ['ab', 'cd', 'ef'],
     'col6': [100, 200, 300]})
# Create a new DataFrame matching col2
# of df2 and col5 of df3.
df1.merge(df3, left_on='col2',
          right_on='col5')
```

## Graphing

```python
import altair as alt

alt.Chart(df2).mark_point().encode(
  x='Country', y='Area', size='Pop',
  color='Birthrate')
```
