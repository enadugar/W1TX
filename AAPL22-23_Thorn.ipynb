```python
import yfinance as yf

# Define the ticker symbol
tickers = ['AAPL']

# Get historical data as dataframe
stock = yf.download(tickers, start='2022-01-01', end='2023-09-08')

# Print the data
print(stock)
```

The table contains historical daily price data for Apple stock (ticker: AAPL) for each trading day between January 3, 2022, and September 7, 2023. The columns in the table are explained as follows:

- **Open**: The price at which Apple stock started on each trading day.
- **High**: The highest price at which Apple stock traded during each trading day.
- **Low**: The lowest price at which Apple stock traded during each trading day.
- **Close**: The price at which Apple stock closed on each trading day.
- **Adj Close**: The adjusted closing price, which accounts for events such as dividends, stock splits, new stock offerings, etc.
- **Volume**: The number of shares of Apple stock that were traded during each trading day.


```python
import matplotlib.pyplot as plt

# Plotting the closing prices
stock['Close'].plot(figsize=(10, 5))
plt.title('Closing Prices of AAPL from 2022-01-03 to 2023-09-07')
plt.ylabel('Closing Price (USD)')
plt.xlabel('Date')
plt.grid(True)
plt.show()
```

```python
import matplotlib.pyplot as plt
import matplotlib.dates as mdates

# Plotting the closing prices
stock['Close'].plot(figsize=(10, 5))

# Set the title and y-axis label
plt.title('Closing Prices of AAPL from (month)-year, (01)-2022 to (09)-2023')
plt.ylabel('Closing Price (USD)')

# Get the current x-axis object and set a custom major formatter
ax = plt.gca()
ax.xaxis.set_major_formatter(mdates.DateFormatter('%m-%y'))

# Display grid and show the plot
plt.grid(True)
plt.show()
```

```python
import matplotlib.pyplot as plt

# Define the date ranges for each cluster
clusters = [
    ('2022-01-01', '2022-06-11'),
    ('2022-06-12', '2023-01-30'),
    ('2023-02-01', '2023-09-07')
]

# Iterate through each cluster and create a separate plot
for i, (start_date, end_date) in enumerate(clusters, 1):
    # Select the data for the current cluster
    cluster_data = stock.loc[start_date:end_date]

    # Plot the closing prices for the current cluster
    cluster_data['Close'].plot(figsize=(10, 5))
    plt.title(f'Closing Prices of AAPL from {start_date} to {end_date}')
    plt.ylabel('Closing Price (USD)')
    plt.xlabel('Date')
    plt.grid(True)
    plt.show()
```

These three 'clusters' chosen by seperate rise and climb actions. These three graphs were generated from the first. A way to break down and use normalized figures. Illustrated by the y-axis.

```python
# Calculate rolling average for one year
one_year_rolling_avg = stock['Close'].rolling(window=252).mean()

# Find the lowest average and its date
lowest_one_year_avg = one_year_rolling_avg.min()
date_lowest_one_year_avg = one_year_rolling_avg.idxmin()

# Display the results
print(f'Lowest One-Year Average: ${lowest_one_year_avg:.2f} on {date_lowest_one_year_avg.strftime("%Y-%m-%d")}')
```

Over the last year, the lowest available price for AAPL stock, was $149.52 ib April, 24th.
