```python
import pandas as pd

def fetch_course_tables(url):
    tables = pd.read_html(url)
    return tables

def extract_department(catnum):
    """Extract the department from the CATNUM."""
    return ''.join([char for char in catnum if char.isalpha()])

def process_table(table):
    # Extracting the department code from CATNUM
    table['department'] = table['CATNUM'].str.extract(r'([A-Z]+)')

    # Counting sections per department
    sections_count = table['department'].value_counts().reset_index()
    sections_count.columns = ['department', 'sections']

    return sections_count.sort_values(by="department")

def compute_course_counts(data):
    counts = defaultdict(lambda: defaultdict(int))
    for _, row in data.iterrows():
        dept = row['department']
        course = row['CATNUM']
        counts[dept][course] += 1
    return counts

# Fetch data
url_2022 = "https://bscintra.buffalostate.edu/msched/202230-all_courses_ug.html"
url_2023 = "https://bscintra.buffalostate.edu/msched/202330-all_courses_ug.html"

tables_2022 = fetch_course_tables(url_2022)
tables_2023 = fetch_course_tables(url_2023)

# Assuming the first table contains the relevant data for both years
df_2022 = process_table(tables_2022[0])
df_2023 = process_table(tables_2023[0])

# Merge the two DataFrames on the "department" column
merged = pd.merge(df_2022, df_2023, on='department', suffixes=('_2022', '_2023'))

# Calculate the difference in section counts
merged['section_difference'] = merged['sections_2023'] - merged['sections_2022']

# Compute counts for both years (this part uses the earlier code)
counts_2022 = compute_course_counts(tables_2022[0])
counts_2023 = compute_course_counts(tables_2023[0])

# Print courses count comparison
comparison_table = []

for dept, courses in counts_2022.items():
    for course, count in courses.items():
        count_2023 = counts_2023.get(dept, {}).get(course, 0)
        comparison_table.append({
            'Dept': dept,
            'Course': course,
            '2022 Count': count,
            '2023 Count': count_2023
        })

# Convert comparison data to a pandas DataFrame
comparison_df = pd.DataFrame(comparison_table)
# ... [Rest of the code remains the same]

# Print courses count comparison
comparison_table = []

# Get the union of all departments and courses from both years
all_departments = set(counts_2022.keys()) | set(counts_2023.keys())

for dept in all_departments:
    all_courses = set(counts_2022.get(dept, {}).keys()) | set(counts_2023.get(dept, {}).keys())

    for course in all_courses:
        count_2022 = counts_2022.get(dept, {}).get(course, 0)
        count_2023 = counts_2023.get(dept, {}).get(course, 0)

        comparison_table.append({
            'Dept': dept,
            'Course': course,
            '2022 Count': count_2022,
            '2023 Count': count_2023
        })

# Convert comparison data to a pandas DataFrame
comparison_df = pd.DataFrame(comparison_table)


def process_all_tables(tables):
    all_data = pd.DataFrame(columns=['department', 'sections'])

    for table in tables:
        processed_data = process_table(table)
        all_data = pd.concat([all_data, processed_data])

    # Group by department and sum the sections
    grouped = all_data.groupby('department').sum().reset_index()
    return grouped

df_2022 = process_all_tables(tables_2022)
df_2023 = process_all_tables(tables_2023)

# Merge the two DataFrames on the "department" column
merged = pd.merge(df_2022, df_2023, on='department', how='outer', suffixes=('_2022', '_2023')).fillna(0)

# Calculate the difference in section counts
merged['section_difference'] = merged['sections_2023'] - merged['sections_2022']

# Display the merged table
merged
```

```python
import matplotlib.pyplot as plt
import seaborn as sns

# Filter out departments with minor changes for better visualization
significant_changes = merged[abs(merged['section_difference']) > 5].sort_values(by="section_difference")

plt.figure(figsize=(15, 10))
sns.barplot(data=significant_changes, y="department", x="section_difference", palette="coolwarm")
plt.title('Difference in Section Counts between 2022 and 2023')
plt.xlabel('Difference in Sections')
plt.ylabel('Department')
plt.show()
```
